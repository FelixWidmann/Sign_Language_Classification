```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
f1_df = pd.read_csv('./f1.csv')
test_loss_df = pd.read_csv('./test_loss.csv')
train_loss_df = pd.read_csv('./train_loss.csv')

# Select runs for plotting
selected_runs_f1 = [
    'upbeat-smoke-18 - weighted_F1',
    'dulcet-paper-16 - weighted_F1',
    'pleasant-wave-21 - weighted_F1',
    'rich-snowflake-20 - weighted_F1'
]

selected_runs_loss = [
    ('upbeat-smoke-18 - train_loss', 'upbeat-smoke-18 - test_loss'),
    ('dulcet-paper-16 - train_loss', 'dulcet-paper-16 - test_loss'),
    ('pleasant-wave-21 - train_loss', 'pleasant-wave-21 - test_loss'),
    ('rich-snowflake-20 - train_loss', 'rich-snowflake-20 - test_loss')
]

# Plot F1 scores
plt.figure(figsize=(12, 6))
for run in selected_runs_f1:
    plt.plot(f1_df['Step'], f1_df[run], label=run.split(' - ')[0])
plt.xlabel('Step')
plt.ylabel('Weighted F1 Score')
plt.title('F1 Score over Steps for Selected Runs')
plt.legend()
plt.grid(True)
plt.show()

# Plot training and test loss in the same graph
plt.figure(figsize=(12, 6))
for train_run, test_run in selected_runs_loss:
    run_name = train_run.split(' - ')[0]
    plt.plot(train_loss_df['Step'], train_loss_df[train_run], label=f'Train {run_name}')
    plt.plot(test_loss_df['Step'], test_loss_df[test_run], label=f'Test {run_name}', linestyle='--')
plt.xlabel('Step')
plt.ylabel('Loss')
plt.title('Training and Test Loss over Steps for Selected Runs')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.grid(True)
plt.show()
```

```python
# Load the CSV files
f1_df = pd.read_csv('./f1.csv')
test_loss_df = pd.read_csv('./test_loss.csv')
train_loss_df = pd.read_csv('./train_loss.csv')

# Select runs for plotting
selected_runs_f1 = [
    'upbeat-smoke-18 - weighted_F1',
    'dulcet-paper-16 - weighted_F1',
    'rich-snowflake-20 - weighted_F1'
]

selected_runs_test_loss = [
    'upbeat-smoke-18 - test_loss',
    'dulcet-paper-16 - test_loss',
    'rich-snowflake-20 - test_loss'
]

selected_runs_train_loss = [
    'upbeat-smoke-18 - train_loss',
    'dulcet-paper-16 - train_loss',
    'rich-snowflake-20 - train_loss'
]

# Manually insert the learning rate and batch size for each run
run_info = {
    'upbeat-smoke-18': {'learning_rate': 0.0001, 'batch_size': 48},
    'rich-snowflake-20': {'learning_rate': 0.001, 'batch_size': },
    'dulcet-paper-16': {'learning_rate': 0.001, 'batch_size': 32}
}

# Plot F1 scores
plt.figure(figsize=(12, 6))
for run in selected_runs_f1:
    run_name = run.split(' - ')[0]
    plt.plot(f1_df['Step'], f1_df[run], label=f'{run_name} (LR: {run_info[run_name]["learning_rate"]}, BS: {run_info[run_name]["batch_size"]})')
plt.xlabel('Step')
plt.ylabel('Weighted F1 Score')
plt.title('F1 Score over Steps for Selected Runs')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.grid(True)
plt.show()

# Plot training loss
plt.figure(figsize=(12, 6))
for run in selected_runs_train_loss:
    run_name = run.split(' - ')[0]
    plt.plot(train_loss_df['Step'], train_loss_df[run], label=f'Train {run_name} (LR: {run_info[run_name]["learning_rate"]}, BS: {run_info[run_name]["batch_size"]})')
plt.xlabel('Step')
plt.ylabel('Training Loss')
plt.title('Training Loss over Steps for Selected Runs')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.grid(True)
plt.show()

# Plot test loss
plt.figure(figsize=(12, 6))
for run in selected_runs_test_loss:
    run_name = run.split(' - ')[0]
    plt.plot(test_loss_df['Step'], test_loss_df[run], label=f'Test {run_name} (LR: {run_info[run_name]["learning_rate"]}, BS: {run_info[run_name]["batch_size"]})', linestyle='--')
plt.xlabel('Step')
plt.ylabel('Test Loss')
plt.title('Test Loss over Steps for Selected Runs')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Adjust legend position
plt.grid(True)
plt.show()
```
